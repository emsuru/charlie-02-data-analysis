## 2. People talk about "villas" and "row houses" in my area. What are they talking about?

In Belgium usually:
  - a villa is **a detached house** (has 4+ facades, you can walk around it)
  - a **semi-detached house** (has 3 facades, you don't need to walk thorugh it to go from your backyard to the front of the house, there is an open side)
  - a row house is **an attached house** (2 facades, it's called a row house because it's in a row of other houses, and has a house 'attached' to it on either side)

 I'll use "number of facades" in my data to segment and show how these types of houses differ in price.

```python
# # check how many null values fo the "Facades" column

print("Total null values in Facades:", df['Facades'].isnull().sum())
# # what is the proportion of null values to total values in the "Facades" column
print("Percentage of null values in total:", f"{(df['Facades'].isnull().sum() / df['Facades'].shape[0]):.2f}")  # 0.3% of the values are null


# drop the rows with null values in 'Facades' & make a copy of the dataframe
df = df.dropna(subset=['Facades']).copy()


# # what are the unique values in the "Facades" column
print("Unique values in Facades:", df['Facades'].unique())

# # what are the unique values in the "Facades" column, only for properties where "PropertyType" is "HOUSE
print("Unique values in Facades for houses:", df[df['PropertyType'] == 'HOUSE']['Facades'].unique())

# #what are the proportions the unique values in the "Facades" column for properties where "PropertySubtype" is "HOUSE"
facades_proportions = df[df['PropertyType'] == 'HOUSE']['Facades'].value_counts(normalize=True).sort_index()
# print each unique value in Facades column and its proportion of total values in the column
#for index, value in facades_proportions.items():
#    print("Percentage for unique value", f'{index}: {value:.4f}')

# how many rows with value over 6 for Facades where PropertyType is HOUSE
print("Number of rows with value over 6 for Facades where PropertyType is HOUSE:",
      df[(df['PropertyType'] == 'HOUSE') & (df['Facades'] > 6)].shape[0])
print("Number of rows with value equal or over 4 for Facades where PropertyType is HOUSE:",
      df[(df['PropertyType'] == 'HOUSE') & (df['Facades'] >= 4)].shape[0])
```

```python
# Create a new column called "HouseType" where we group all properties of "PropertyType" "HOUSE" into three groups by the number of "Facades": "Attached house" : 2 facades, "Semi-attached house": 3 facades, "Detached house": 4 facades or more:
df['HouseType'] = np.where(df['Facades'] == 2, 'Attached_House', np.where(df['Facades'] == 3, 'SemiDetached_House', 'Detached_House'))

#check new column HouseType, print unique values and how many of each unique value in teh dataset
print("Unique values in HouseType:", df['HouseType'].unique())
print("How many houses in teh dataset by house type:", df['HouseType'].value_counts())
```

```python
#Median price (better than mean to not be influenced by outliers)for a 'Detached house' in the region of Flanders?
flanders_detached_houses = df[(df['Region'] == 'FLANDERS') & (df['HouseType'] == 'Detached_House')]
print(f"The median price for a detached house in the region of Flanders is: {int(flanders_detached_houses['Price'].median()):,}€")


# Price for a detached house in the district of Gent
gent_detached_houses = df[(df['District'] == 'Gent') & (df['HouseType'] == 'Detached_House')]
print(f"The median price for a detached house in the district of Gent is: {int(gent_detached_houses['Price'].median()):,}€")

# Price for a semi-detached house in Flanders
flanders_semi_detached_houses = df[(df['Region'] == 'FLANDERS') & (df['HouseType'] == 'SemiDetached_House')]
print(f"The median price for a semi-detached house in the region of Flanders is: {int(flanders_semi_detached_houses['Price'].median()):,}€")

# Price for an attached house (aka a "row" house) in Flanders
flanders_attached_houses = df[(df['Region'] == 'FLANDERS') & (df['HouseType'] == 'Attached_House')]
print(f"The median price for an attached house in the region of Flanders is: {int(flanders_attached_houses['Price'].median()):,}€")
```

```python
# Create a new column 'PropertyCategory' that combines 'PropertyType' and 'HouseType'
df['PropertyCategory'] = np.where(df['PropertyType'] == 'APARTMENT', 'Apartment', df['HouseType'])

# the default graph has ugly colours

# Define custom palette to make it look a bit better
custom_palette = {'Detached_House': '#1f77b4',  # Blue shade
                  'SemiDetached_House': '#aec7e8',  # Lighter blue shade
                  'Attached_House': '#3498db',  # Another blue shade
                  'Apartment': '#ff7f0e'}  # Orange for apartments

# Calculate median prices for each combination of 'Region' and 'PropertyCategory'
median_prices = df.groupby(['Region', 'PropertyCategory'])['Price'].median().reset_index()

# Create the grouped bar chart with the custom palette
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Region', y='Price', hue='PropertyCategory', data=median_prices, palette=custom_palette)

plt.title('Median Prices by Property Category and Region')
plt.xlabel('Region')
plt.ylabel('Median Price (Million €)')
plt.xticks(rotation=45)
plt.legend(title='Property Category')

# Format y-axis to show prices in million euros instead of scientific notation
barplot.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.2f}M'))

# Show the plot
plt.tight_layout()
plt.show()

# save graph as png in the graphs folder
barplot.figure.savefig('graphs/median_prices_by_property_category_and_region.png')
```

## 4. What are the market segments and their proportions by price?

---


- low-end (budget)
- mid-end (standard)
- high-end (luxury)

E.g.: 60% of the market is in the 200k-400k price segment. 
Maybe show data only on castles and very high-end properties. Could be fun to see.


```python
# TO EXPLORE LATER, MY TIME IS UP and I'm tired

# Ensure that 'Price' is in numeric format
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Define bins and labels for all properties
bins = [0, 200000, 400000, 600000, 800000, 1000000, float('inf')]
labels = ['0-200k', '200k-400k', '400k-600k', '600k-800k', '800k-1M', '>1M']

# Create 'Price_Segment' for all properties
df['Price_Segment'] = pd.cut(df['Price'], bins=bins, labels=labels, right=False)

# Calculate the proportion of each price segment
price_segment_proportions = df['Price_Segment'].value_counts(normalize=True).sort_index()

# Plot the proportions
plt.figure(figsize=(10, 6))
bars = sns.barplot(x=price_segment_proportions.index, y=price_segment_proportions.values, palette="Blues_d")

plt.title('Proportion of Properties by Price Segment')
plt.xlabel('Price Segment')
plt.ylabel('Proportion')
plt.xticks(rotation=45)  # Rotate labels to improve readability

# Annotate each bar with the percentage value
for bar in bars.patches:
    # The text annotation for each bar should be its height (proportion) formatted as a percentage
    bars.annotate(format(bar.get_height(), '.1%'),  # Format the proportion as a percentage with one decimal place
                   (bar.get_x() + bar.get_width() / 2, bar.get_height()),  # Position for the text (x, y)
                   ha='center',  # Center the text horizontally
                   va='center',  # Center the text vertically within the bar
                   size=10,  # Font size
                   xytext=(0, 8),  # Position offset for the text (to be slightly above the bar)
                   textcoords='offset points')  # Use offset points for positioning the text

plt.show()

#save graph as png in graphs folder
bars.figure.savefig('graphs/proportion_of_properties_by_price_segment.png')
```

```python
# Calculate the count of each combination of 'Price_Segment' and 'Type'
segment_type_counts = df.groupby(['Price_Segment', 'PropertyType']).size().reset_index(name='counts')

# Calculate the total count of properties in each 'Price_Segment'
segment_counts = df['Price_Segment'].value_counts().reset_index()
segment_counts.columns = ['Price_Segment', 'total_counts']

# Merge to get total counts in each segment for normalization
segment_type_counts = pd.merge(segment_type_counts, segment_counts, on='Price_Segment')

# Calculate the proportion of each type within each price segment
segment_type_counts['proportion'] = segment_type_counts['counts'] / segment_type_counts['total_counts']

# Now plot using sns.barplot with the 'hue' parameter
custom_palette = {'HOUSE': 'green', 'APARTMENT': 'blue'}
plt.figure(figsize=(12, 6))
sns.barplot(data=segment_type_counts, x='Price_Segment', y='proportion', hue='PropertyType', palette=custom_palette)

plt.title('Proportion of Property Types by Price Segment')
plt.xlabel('Price Segment')
plt.ylabel('Proportion')
plt.xticks(rotation=45)

plt.show()
```
